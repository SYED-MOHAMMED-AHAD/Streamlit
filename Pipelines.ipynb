```python
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PowerTransformer
from sklearn.compose import ColumnTransformer
import streamlit as st
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  RidgeClassifier
```

```python
df = pd.read_csv("E:\Machine Learning Projects\BA PROJECT\PART-2\cleaned_data.csv")
```

```python
x = df.drop('booking_complete',axis = 1)
```

```python
x.shape
```

```python
y = df['booking_complete']
y.shape
```

```python
x.select_dtypes(exclude = ['int', 'float']).columns
```

```python
cont_pl = Pipeline(steps=[
    ('scaler', StandardScaler()) ])

cat_pl = Pipeline(steps=[
    ('onehot', OneHotEncoder(drop='first')) ])



ct = ColumnTransformer(
    transformers=[
        ('Continuous transformer', cont_pl, ['num_passengers', 'purchase_lead', 'length_of_stay', 'flight_hour',
                                             'flight_day', 'wants_extra_baggage', 'wants_preferred_seat',
       'wants_in_flight_meals', 'flight_duration', 'all_services_opted']),
        ('Cat Nominal transformer', cat_pl, ['sales_channel', 'trip_type', 'route', 'booking_origin',
       'parts_of_the_day', 'purchase_lead_bin']) 
    ],
    remainder='passthrough'
)

final_pl = Pipeline(steps=[
    ('ct', ct),  # Apply the ColumnTransformer
    ('model', RidgeClassifier(alpha=1.0))  # the model
])

final_pl
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Fit the pipeline on training data
final_pl.fit(X_train,y_train)
```

```python
import pickle
pickle.dump(final_pl,open('estimator1.pkl','wb'))
predict = pickle.load(open('estimator1.pkl','rb'))
```
